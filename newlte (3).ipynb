```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import dash
from dash import dash_table, html
```

```python
df= pd.read_csv("C:/Users/User/Downloads/employee_data.csv")
df
```

```python
df
```

```python
df= df.fillna("Arun")
df
```

```python
df["Age"]=df["Age"].replace("Arun","28.0")
df
```

```python
df["Salary"]=df["Salary"].replace("Arun","65000")
df
```

```python
df['Salary']= df['Salary'].astype('int')
df
```

```python
df= df.drop_duplicates()
df
```

```python
data = {
    "Name": ["Rajesh","Nasir","Mathiew","Gilbert","Kevin","Navin","Arun"],
    "Salary": [60000,65000, 58000,60000,61000, 72000,64000],
     "Age": [29,31,28,29,28,35,40],
    "Department": ['Ops', 'IT Tech','Support', 'Ops','Support','IT','Support'],  
}

df = pd.DataFrame(data)
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H2("Employee data"),
    dash_table.DataTable(
        id='Employee data-grid',
        columns=[{"name": i, "id": i} for i in df.columns],
        data=df.to_dict('records'),
        style_table={'overflowX': 'auto'},
        style_cell={
            'textAlign': 'center',
            'padding': '10px',
            'fontFamily': 'Arial'
        },
        style_header={
            'backgroundColor': '#f2f2f2',
            'fontWeight': 'bold'
        },
       
    )
])

if __name__ == '__main__':
    app.run(debug=True)
```
